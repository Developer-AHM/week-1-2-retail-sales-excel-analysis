```python
# KPI Verification
print("Total Sales:", df["Total Amount"].sum())
print("Total Transactions:", df["Transaction ID"].count())
print("Average Sales:", df["Total Amount"].mean())
print("Total Quantity Sold:", df["Quantity"].sum())
print("Highest Transaction:", df["Total Amount"].max())
print("Lowest Transaction:", df["Total Amount"].min())
```
